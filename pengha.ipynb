```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

```python
def ImgNormalization(img):
    resizedIMG = cv2.resize(img, dsize=(227,227))
    RGBIMG = cv2.cvtColor(resizedIMG, cv2.COLOR_BGR2RGB)

    B, G, R= cv2.split(RGBIMG)
    proR = np.float64(R) / 255.0
    proG = np.float64(G) / 255.0
    proB = np.float64(B) / 255.0
    result = cv2.merge((proB, proG, proR))
    return result
```

```python
def Load_FileName(path): # 해당 폴더에 있는 모든 파일 이름들 불러오는 코드
    folder_list = os.listdir(path)
    XData = []
    number = []
    p = 0
    for i in folder_list:
        file_list = os.listdir(path + '/' + i)
        number.append(i)
        for j in file_list:
            print(path + '/' + i + '/' + file_list[p])
            c = cv2.imread(path + '/' + i + '/' + file_list[p])
            result = ImgNormalization(c)
            XData.append(result) # Data_set 리스트 안에 사진 정보를 차례대로 넣는다
            XData.append(cv2.flip(result, 1)) # 좌우 반전
            p += 1
        p = 0
    print(number)
    print(np.array(XData))
    return np.array(XData)# 리스트를 배열화 시킨다.

def Make_Y(path, Y_list):
    folder_name = os.listdir(path)
    Y = []
    number = []
    p = 0
    for i in folder_name:
        file_name = os.listdir(path+'/'+ i)
        number.append(i)
        for j in file_name:
            Y.append(Y_list[p])
            Y.append(Y_list[p])
        p += 1
    print(number)
    return np.array(Y)
```

```python
X = Load_FileName('C:/Users/asdew32/Desktop/train')
# 테스트 데이터셋 가져오기

ValX = Load_FileName('C:/Users/asdew32/Desktop/test2/validation')
# 검증을 위한 데이터셋

Y = Make_Y('C:/Users/asdew32/Desktop/train', ['A', 'B', 'C', 'D', 'E', 'F']) # 답지 데이터셋 만들기
ValY = Make_Y('C:/Users/asdew32/Desktop/test2/validation', ['A', 'B', 'C', 'D', 'E', 'F']) # 검증을 위한 답지 데이터

# 왕관 펭귄 : 쇠푸른 펭귄 A, 왕관 펭귄 B, 황제 펭귄 C, 젠투 펭귄 D, 줄무늬 펭귄 E

Y = pd.get_dummies(Y) # 원 핫 인코딩
ValY = pd.get_dummies(ValY)
print(X.shape)
print(Y.shape)
print(ValX.shape)
print(ValY.shape)
```

```python
# 5개의 컨볼루션 3Dense층
studyX = tf.keras.layers.Input(shape=[227, 227, 3])

H = tf.keras.layers.Conv2D(96, kernel_size=11, strides=(4, 4), activation="relu")(studyX)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Conv2D(256, kernel_size=5, padding="same", activation="relu")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(H)

H = tf.keras.layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(H)

H = tf.keras.layers.Conv2D(256, kernel_size=3, padding="same", activation="relu")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(4096, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01))(H) # 가중치 규제 L2 적용
H = tf.keras.layers.Dropout(0.5)(H)
H = tf.keras.layers.Dense(4096, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01))(H) # 가중치 규제 L2 적용 (Dense 층에만 적용하는 것이 좋다고 GPT에게 들음)
H = tf.keras.layers.Dropout(0.5)(H)
studyY = tf.keras.layers.Dense(6, activation="softmax")(H)

model = tf.keras.models.Model(studyX, studyY)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, Y, epochs = 120, validation_data = (ValX, ValY), batch_size = 10)
```

```python
end = 100 + 1
epoch = np.arange(1, end) # 1 ~ 1+n

plt.plot(epoch, history.history["loss"], epoch, history.history["val_loss"], "r-")

plt.show()
```

```python
plt.plot(epoch, history.history["accuracy"], epoch, history.history["val_accuracy"], 'r-')

plt.show()
```
